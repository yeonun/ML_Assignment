# Import packages numpy for calculating and matplotlib for drawing graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define a differentiable function that maps from real number to real number.

## Define $f(x) = x*cosx$ as func(x)

```python
def func(x):
    f = np.cos(x)*x
    return f
```

# Define a domain of the function.

## Domain : $ -5 < x < 5 $

```python
x = np.arange(-5,5,0.1)
```

# Plot the function.

```python
f = func(x)
plt.figure(1)
plt.plot(x,f,'k',label="function")
```

# Select a point within the domain

## point $x$ : -3,4

```python
p1 = -3
p2 = 4
```

# Mark the selected point on the function

```python
def origin_plot():
    y1 = func(p1)
    y2 = func(p2)
    plt.scatter(p1,y1,c='r',s=35)
    plt.scatter(p2,y2,c='g',s=35)
    plt.plot(x,f,'k',label="function")
    
origin_plot()
```

# Define the first-order Taylor approximation at the selected point

## Define $f'(x) = -x*sinx + cosx$ as derivate function d_func(x)

```python
def d_func(x):
    df = - x*np.sin(x) +np.cos(x)
    return df
```

## Define Tylor Approximation $f(a) + f'(a)(x-a)$

```python
def tylor(a,x):
    result = func(a) + d_func(a)*(x-a)
    return result
```

# Plot the Taylor approximation with the same domain of the original function.

```python
origin_plot()

df1 = tylor(p1,x)
df2 = tylor(p2,x)
plt.plot(x, df1, 'r')
plt.plot(x, df2, 'g')
```
